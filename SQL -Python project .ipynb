```python
%%sql SELECT 
    "Elementary, Middle, or High School" AS School_Type,
    ROUND(AVG(SAFETY_SCORE)) AS Avg_Safety
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY School_Type 
```

```python
%%sql select Track_Schedule,round(AVG(SAFETY_SCORE))
from CHICAGO_PUBLIC_SCHOOLS
group by Track_Schedule
```

```python
%%sql select COMMUNITY_AREA_Name,COMMUNITY_AREA_NUMBER
   from  CENSUS_DATA
    where HARDSHIP_INDEX in ( select Max(HARDSHIP_INDEX)from  CENSUS_DATA)
```

```python
#Use a sub-query to determine the Community Area Name with most number of crimes?
```

```python
%%sql select  COMMUNITY_AREA_NUMBER , count(*) from CHICAGO_CRIME_DATA
    group by COMMUNITY_AREA_NUMBER 
    order by count(*) DESC
    limit 1
```

```python
%%sql select COMMUNITY_AREA_NAME from CENSUS_DATA
where COMMUNITY_AREA_NUMBER in (select  COMMUNITY_AREA_NUMBER from CHICAGO_CRIME_DATA
    group by COMMUNITY_AREA_NUMBER 
    order by count(*) DESC
    limit 1)
```
